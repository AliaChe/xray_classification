```python
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image
```

```python
data_dir = Path("../data/raw/chest_xray")

train_dir = data_dir / "train"
test_dir = data_dir / "test"
val_dir = data_dir / "val"
```

```python
for split in [train_dir, test_dir, val_dir]:
    print(f"\n{split.name.upper()}")

    for class_dir in split.iterdir():
        if class_dir.is_dir():
            n_images = len(list(class_dir.glob("*")))
            print(f"{class_dir.name}: {n_images}")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10, 6))

classes = ["NORMAL", "PNEUMONIA"]

for row, class_name in enumerate(classes):
    image_paths = list((train_dir / class_name).glob("*"))

    for col in range(3):
        img = Image.open(image_paths[col])

        axes[row, col].imshow(img, cmap="gray")
        axes[row, col].set_title(class_name)
        axes[row, col].axis("off")

plt.tight_layout()
plt.show()
```

```python
sizes = []

image_paths = list((train_dir / "NORMAL").glob("*"))[:100]

for path in image_paths:
    img = Image.open(path)
    sizes.append(img.size)

print(sizes[:10])
```
